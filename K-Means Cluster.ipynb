```python

import numpy as np
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.cross_validation import StratifiedKFold
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
```

```python
iris = datasets.load_iris()
X,y = iris.data[:,:3], iris.target
#z = iris.data[:,:]
```

```python
y
```

```python
num_clusters = 8
model = KMeans(n_clusters=num_clusters)
model.fit(X)
```

```python
labels = model.labels_
cluster_centers = model.cluster_centers_
print (cluster_centers)
#print (labels)
```

```python
plt.scatter(X[:,0], X[:,1],c=labels.astype(np.float))
plt.scatter(cluster_centers[:,0], cluster_centers[:,1], c = np.arange(num_clusters), marker = '^', s = 250)
plt.show()
plt.scatter(X[:,0], X[:,1],c=np.choose(y,[0,2,1]).astype(np.float))
plt.show()
```
